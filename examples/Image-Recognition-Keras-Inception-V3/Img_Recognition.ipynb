# Image Recogniton With Keras + Pretrained model Inception V3

### Load required libraries

```python
from keras.applications.inception_v3 import InceptionV3
from keras.applications.inception_v3 import preprocess_input
from keras.applications.inception_v3 import decode_predictions
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

### Load the InceptionV3 model

```python
model = InceptionV3(weights='imagenet', include_top=True)
```

### Print Structure and Weights, If you want

```python
print("Structure: ", model.summary())
print("Weights: ", model.get_weights())
```

### Predict images using InceptionV3 model

```python
for i in range(0, 15):
    img_path = 'testimg/%s.jpg' % (i)

    img = image.load_img(img_path, target_size=(299, 299))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis = 0)
    x = preprocess_input(x)

    features = model.predict(x)
    print(decode_predictions(features, top = 3))

    plt.imshow(image.load_img(img_path))
    plt.show()
```
